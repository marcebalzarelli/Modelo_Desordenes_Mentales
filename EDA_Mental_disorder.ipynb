# Transformación de Datos y Análisis de Datos

```python
data_md = data_md.rename(columns={'ag+1:629e':'age'})
data_md= data_md.rename(columns={'having.trouble.in.sleeping':'trouble.sleeping'})
data_md = data_md.rename(columns={'having.trouble.with.work':'trouble.with.work'})
data_md = data_md.rename(columns={'having.nightmares':'nightmares'})

data_md.set_index(['age'])
```

```python
edades = data_md['age']

promedio_edad = edades.mean()#Calculo promedio de edades
print(promedio_edad)
```

Observo que el promedio de edades que tenemos en nuestros datos es de aproximadamente 22 años y la gráfica de distribución de las edades me muestra que la mayoria de pacientes se encuentran entre los 12 y 30 años.

```python
sns.histplot(data_md['age'], kde=True)  # Grafica de distribución de las edades

plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución de Edades')

plt.show()
```

```python
Disorders = data_md.Disorder.unique()# Obtengo lista de trastornos únicos
Disorders
```

```python
data_md.Disorder = data_md.Disorder.str.replace('psychotic deprission','psychot depresn')# Corrijo los errores ortográficos y convierto el array en una lista
data_md.Disorder = data_md.Disorder.str.replace('sleeping disorder','sleep disord')
data_md.Disorder = data_md.Disorder.str.replace('anexiety','anxiety')
data_md.Disorder = data_md.Disorder.str.replace('eating disorder','ED')

Disorders = data_md.Disorder.unique().tolist()
```

```python
plt.figure(figsize=(10,6))# Gráfica de Edad media de aparición de diversos trastornos
sns.barplot(data = data_md, x='Disorder', y='age', estimator=np.mean).set_title('Edad media de aparición de diversos trastornos con un IC del 95%.')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))  #Graficaa de Cantidad de pacientes por tipo de trastorno
sns.countplot(data=data_md, x='Disorder', palette='Set1')  
plt.xlabel('Tipo de Trastorno')
plt.ylabel('Cantidad de Pacientes')
plt.title('Cantidad de Pacientes por Tipo de Trastorno')
plt.xticks(rotation=90)#Roto etiquetas
plt.tight_layout() 
plt.show()
```

```python
byDisorder = data_md.groupby('Disorder').sum()# Agrupo los datos por trastorno, para ver grupos y tendencias.
byDisorder.drop('age', axis=1, inplace = True)
byDisorder['incidencias'] = incidencias
byDisorder
```

```python
disorder_values = ["ADHD", "ASD", "ED", "Loneliness", "MDD", "OCD", "PDD", "PTSD", "anxiety", "bipolar", "psychot depresn", "sleep disord"]
byDisorder['Disorder'] = disorder_values
nombre_archivo = 'porDesorden.csv'
byDisorder.to_csv(nombre_archivo, index=False)  

print(f'El archivo {nombre_archivo} se ha guardado con éxito.')
```

```python
d = byDisorder['incidencias']# Divido por la tasa de incidencia de cada trastorno para hallar la prevalencia
byDisorderRate = pd.concat([byDisorder, byDisorder.div(d, axis=0).assign(incidences=d).rename(lambda x: x+' rate')])
byDisorderRate.drop(byDisorderRate.index[0:12], axis=0, inplace=True)
byDisorderRate
# byDisorder
```
